```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from random import randint
import scipy.integrate as spi
import numpy as np
import pylab as pl
```

```python
N = 100

import os
 
def saveplot(path, ext='png', close=True, verbose=True):
    """Save a figure from pyplot.
    Parameters
    ----------
    path : string
    The path (and filename, without the extension) to save the
    figure to.
    ext : string (default='png')
    The file extension. This must be supported by the active
    matplotlib backend (see matplotlib.backends module). Most
    backends support 'png', 'pdf', 'ps', 'eps', and 'svg'.
    close : boolean (default=True)
    Whether to close the figure after saving. If you want to save
    the figure multiple times (e.g., to multiple formats), you
    should NOT close it in between saves or you will have to
    re-plot it.
    verbose : boolean (default=True)
    Whether to print information about when and where the image
    has been saved.
    """
    # Extract the directory and filename from the given path
    directory = os.path.split(path)[0]
    filename = "%s.%s" % (os.path.split(path)[1], ext)
    if directory == '':
        directory = '.'

    # If the directory does not exist, create it
    if not os.path.exists(directory):
        os.makedirs(directory)

    # The final path to save to
    savepath = os.path.join(directory, filename)

    if verbose:
        print("Saving figure to '%s'..." % savepath),

    # Actually save the figure
    plt.savefig(savepath)
    # Close it
    if close:
        plt.close()

    if verbose:
        print("Done") 


def SIR_cure_model(lamb, gamma, ND, S0, L0, I0, R0, P0, D0, vac, cure, alpha):
    TS = 1.0
    INPUT = (S0, L0, I0, R0, P0, D0)

    def SIR(INP,t):  
        '''The main set of equations'''
        Y=np.zeros((6))
        V = INP    
        Y[0] = ((- lamb * V[0] * V[2]) - vac*V[0])/N
        Y[1] = ((lamb * V[0] * V[2]))/N - beta * V[1]
        Y[2] = beta*V[1] - ((gamma + cure) * V[2]) - alpha*V[2]
        Y[3] = (gamma + cure) * V[2]
        Y[4] = (vac*V[0])/N
        Y[5] = alpha*V[2]
        return Y   # For odeint

    t_start = 0.0; t_end = ND; t_inc = TS
    t_range = np.arange(t_start, t_end+t_inc, t_inc)
    RES = spi.odeint(SIR,INPUT,t_range)

    #Ploting
    #pl.subplot(211)
    #pl.plot(RES[:,0], '--g', label='Susceptibles')
    #pl.plot(RES[:,1], '-c', label='Latent')
    #pl.plot(RES[:,3], '-k', label='Contracted')
    #pl.plot(RES[:,4], '-b', label = 'Prevented')
    #pl.plot(RES[:,2], '-r', label='Infectious')
    #pl.plot(RES[:,5], '-y', label = 'Dead')
    #pl.legend(loc= (1,0))
    #pl.title('Program_2_1.py')
    #pl.xlabel('Time (Days)')
    #pl.ylabel('Population (%)')
    return RES.item(ND, 0), RES.item(ND, 1), RES.item(ND, 2), RES.item(ND, 3), RES.item(ND, 4), RES.item(ND, 5);

def SEIR_cure_model(S0, L0, I0, R0, P0, D0, lamb, gamma, vac, cure, alpha, beta, t_f, t_i, t_step):
    INPUT = (S0, L0, I0, R0, P0, D0)

    def SIR(INP,t):  
        '''The main set of equations'''
        Y=np.zeros((6))
        V = INP    
        Y[0] = ((- lamb * V[0] * V[2]) - vac*V[0])/N
        Y[1] = ((lamb * V[0] * V[2]))/N - beta * V[1]
        Y[2] = beta*V[1] - ((gamma + cure) * V[2]) - alpha*V[2]
        Y[3] = (gamma + cure) * V[2]
        Y[4] = (vac*V[0])/N
        Y[5] = alpha*V[2]
        return Y   # For odeint

    t = np.arange(t_i, t_f+t_step, t_step)
    RES = spi.odeint(SIR,INPUT,t)
    return RES

def SEIR_plot(RES):
    pl.subplot(211)
    pl.plot(RES[:,1], '-c', label='Exposed')
    pl.plot(RES[:,2], 'r-.', label='Infectious')
    pl.plot(RES[:,3], ':', label='Contracted')
    pl.plot(RES[:,5], 'k+', label = 'Dead')
    pl.legend(loc= (1,0))
    pl.title('SEIR model with fatalities and vaccinations (1)')
    pl.xlabel('Time (Days)')
    pl.ylabel('Population (%)')
    pl.xlim(0, 100)
    pl.ylim(0, 20)
    pl.show()
    
    pl.subplot(211)
    pl.plot(RES[:,0], 'g', label='Susceptibles')
    pl.plot(RES[:,3], ':', label='Contracted')
    pl.plot(RES[:,4], '--b', label = 'Prevented')
    pl.legend(loc= (1,0))
    pl.title('SEIR model with fatalities and vaccinations (2)')
    pl.xlabel('Time (Days)')
    pl.ylabel('Population (%)')
    
def SEIR_time_delay(S0, L0, I0, R0, P0, D0, lamb, gamma, vac, cure, alpha, beta, t_f, t_i, t_step, time_delay, FINAL):
    t_d = time_delay
    ND = t_d
    
    O = SEIR_cure_model(S0, L0, I0, R0, P0, D0, lamb, gamma, 0.0, 0.0, alpha, beta, t_d, t_i, t_step)
    SNew, LNew, INew, RNew, PNew, DNew = [O.item(t_d, 0), O.item(t_d, 1), O.item(t_d, 2), O.item(t_d, 3),O.item(t_d, 4), O.item(t_d, 5)]
    
    A = SEIR_cure_model(SNew, LNew, INew, RNew, PNew, DNew, lamb, gamma, vac, cure, alpha, beta, t_f, t_d, t_step)
    
    S = O[:,0].tolist() + A[:,0].tolist()
    E = O[:,1].tolist() + A[:,1].tolist()
    I = O[:,2].tolist() + A[:,2].tolist()
    R = O[:,3].tolist() + A[:,3].tolist()
    P = O[:,4].tolist() + A[:,4].tolist()
    D = O[:,5].tolist() + A[:,5].tolist()
    RES = [S, E, I, R, P, D]
    if FINAL == False:
        return RES
    if FINAL == True:
        return S[t_d], E[t_d], I[t_d], R[t_d], P[t_d], D[t_d], 
    

def SEIR_plot_delay(RES, t_f, title, extra, introduction):
    pl.subplot(211)
    x = np.arange(0, t_f+2, 1)
    
    pl.plot(x, RES[1], '-c', label='Exposed ($E$)')
    pl.plot(x, RES[2], 'r-.', label='Infectious ($I$)')
    pl.plot(x, RES[3], ':', label='Recovered ($R$)')
    pl.plot(x, RES[5], 'k+', label = 'Dead ($D$)')
    pl.legend(loc= (1,0))
    pl.title(title)
    pl.xlabel('Time (Days)')
    pl.ylabel('Population (%)')
    pl.xlim(0, t_f)
    pl.ylim(0, 2)
    pl.axis('tight')
   #pl.annotate('Intervention', xy=(introduction, 0), xytext=(introduction - 3.4, -.3),
    #            arrowprops=dict(facecolor='black', arrowstyle = "->"))

    
    pl.show()
    if extra == True:
        pl.subplot(211)
        pl.plot(x, RES[0], 'g', label='Susceptibles ($S$)')
        pl.plot(x, RES[3], ':', label='Recovered ($R$)')
        pl.plot(x, RES[4], '--b', label = 'Prevented ($P$)')
        pl.legend(loc= (1,0))
        pl.xlim(0, t_f)
        pl.title(title)
        pl.xlabel('Time (Days)')
        pl.ylabel('Population (%)')
```

```python
lamb = .2
gamma = .1
ND = 100
S0, L0, I0, R0, P0, D0 = [N, 0.00, .01*N, 0.0, 0.0, 0.0]
vac = 0.00
cure = 0.0
alpha = 1.0/6.5
beta = 1.0/10.1

x = np.arange(0, 2, .05)
dead_vac = []
dead_cure = []

for i in x:
    vac = i
    result = SIR_cure_model(lamb, gamma, ND, S0, L0, I0, R0, P0, D0, vac, 0.0, alpha);
    dead_vac.append(result[5])
for i in x:
    cure = i
    result = SIR_cure_model(lamb, gamma, ND, S0, L0, I0, R0, P0, D0, 0.0, cure, alpha);
    dead_cure.append(result[5])
```

```python
plt.plot(x, dead_vac, '--b', label = 'Vaccine Alone')
plt.plot(x, dead_cure, 'r', label = 'Cure Alone')
plt.title('Medicine Effectiveness Comparison')
plt.xlabel(r'Value of $v$ and $c$')
plt.ylabel('Deaths (% of Population)')
plt.legend(loc = 1)
plt.ylim(0, 3)
plt.xlim(0, 1.95)
#saveplot('Cure_vs_Vaccine', 'png')
plt.show()
```
