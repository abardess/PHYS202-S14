```python
import scipy.integrate as spi
import numpy as np
import pylab as pl


lamb, gamma, ND, S0, I0, R0, P0, D0, vac, cure, alpha = [.2, .1, 100.0, 1-1e-6, 1e-6, 0.0, 0.0, 0.0,  0.01, 0.01, 0.07]


def SIR_cure_model(lamb, gamma, ND, S0, I0, R0, P0, D0, vac, cure, alpha):
    TS = 1.0
    INPUT = (S0, I0, R0, P0, D0)

    def SIR(INP,t):  
        '''The main set of equations'''
        Y=np.zeros((5))
        V = INP    
        Y[0] = - lamb * V[0] * V[1] - vac*V[0]
        Y[1] = lamb * V[0] * V[1] - (gamma + cure) * V[1] - alpha*V[1]
        Y[2] = (gamma + cure) * V[1]
        Y[3] = vac*V[0]
        Y[4] = alpha*V[1]
        return Y   # For odeint

    t_start = 0.0; t_end = ND; t_inc = TS
    t_range = np.arange(t_start, t_end+t_inc, t_inc)
    RES = spi.odeint(SIR,INPUT,t_range)

    #Ploting
    pl.subplot(211)
    pl.plot(RES[:,0], '-g', label='Susceptibles')
    pl.plot(RES[:,2], '-k', label='Recovered')
    pl.plot(RES[:,3], '-b', label = 'Prevented')
    pl.plot(RES[:,1], '-r', label='Infectious')
    pl.plot(RES[:,2], '-k', label='Recovered')
    pl.plot(RES[:,4], '-y', label = 'Dead')
    pl.legend(loc= (1,0))
    pl.title('Program_2_1.py')
    pl.xlabel('Time')
    pl.ylabel('Susceptibles and Recovereds')
    pl.subplot(212)
    pl.plot(RES[:,1], '-r', label='Infectious')
    pl.plot(RES[:,2], '-k', label='Recovered')
    pl.plot(RES[:,4], '-y', label = 'Dead')
    pl.xlabel('Time')
    pl.ylabel('Infectious')
    pl.show()
    return RES.item(ND, 0), RES.item(ND, 1), RES.item(ND, 2), RES.item(ND, 3), RES.item(ND, 4,)

SIR_cure_model(lamb, gamma, ND, S0, I0, R0, P0, D0, vac, cure, alpha);
```

```python
center = [0, 0]
outbreak = [100, 100]
delivery_speed = 1
r = np.sqrt(outbreak[0]**2 + outbreak[1]**2)
travel_time = r/delivery_speed
vac = 0.00
cure = 0.00

SNew, INew, RNew, PNew, DNew = SIR_cure_model(lamb, gamma, travel_time, S0, I0, R0, P0, D0, vac, cure, alpha)

vac = 0.03
cure = 0.10

SIR_cure_model(lamb, gamma, ND, SNew, INew, RNew, PNew, DNew, vac, cure, alpha);
```
