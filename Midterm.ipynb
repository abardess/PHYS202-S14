Question 2 Solution:

```python
from ipythonblocks import BlockGrid
grid = BlockGrid(8,8, fill = (123, 234, 123))
BLUE = (0, 150, 150)
grid[:, 1] = BLUE 
grid[0, 2:4] = BLUE 
grid[3:5, 2:4] = BLUE 
grid[7, 2:4] = BLUE 
grid[1:3, 4] = BLUE 
grid[5:7, 4] = BLUE
grid.show()
```
