```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from random import randint
```

```python
def random_path(iterations, steps, running_average = False):
    """Simulates a particle moving throught [iterations] iterations of a random path of [steps] number of steps
    and returns a plot of all paths taken as well as a histogram of the particle's final positions. Also returns
    the average position of the particle across all iterations. Can also be set to give a running average of each
    iteration--this is not recoomended"""
    x_n = []                        #record the last(nth) position of our particle
    xave_total = 0
    for n in range(iterations):     #simulate a 1000 step random path
        x = 0                       #initial position
        path = []                   #record path taken

        for i in range(steps):   #randomly takes a step forward or backward
            path.append(x)
            s = randint(0, 2)
            if s == 1:
                x += 1
            if s == 0:
                x += -1
        x_n.append(x)           #record the last(nth) position of our particle
        y = np.arange(0, 1000, 1)
        plt.plot(y, path, color = numpy.random.rand(3,1))        #plot position versus step
        xave = float(x)/float(steps)                             #compute average x for this iteration
        xave_total += xave #compute running average
        if running_average == True:
            print 'Iteration', n, 'Average:', xave               #dislpay average for each iteration (W/BIG NUMBERS = NOT GOOD IDEA)
    xave_total = float(xave_total)/float(iterations)             #compute total average x
    print 'Average x for all iterations:', xave_total
    plt.show()                                                   #display all position plots

    n, bins, patches = plt.hist(x_n, 200, normed = 1)            #create histogram of each final position of our particle
    plt.grid();
```

```python
random_path(4000, 1000)
```
