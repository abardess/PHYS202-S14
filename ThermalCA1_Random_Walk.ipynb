```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from random import randint
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
def random_step():
    """Generates a unit step forward or backward at random """
    s = randint(0, 1)
    if s == 1:
        return 1
    if s == 0:
        return -1         

def random_path(iterations, steps):
    """Simulates a particle moving throught [iterations] iterations of a random path of [steps] number of unit steps
    and returns a plot of all paths taken as well as a histogram of the particle's final positions. Also returns the
    average position of the particle across all iterations."""
    r_n = []                                    #record the last(nth) position of the simulated particle
    rave_total = [0,0,0]
    xpaths = []
    ypaths = []
    zpaths = []
    for n in range(iterations):                 #simulate [iterations] iterations of a randomized path
        r = [0,0,0]                             #initial position
        xpath = []; ypath = []; zpath = []      #record path taken in all dimensions
        for i in range(steps):                  #randomly takes a unit step forward or backward for [steps] number of steps
            xpath.append(r[0])
            ypath.append(r[1])
            zpath.append(r[2])
            r[0] += random_step()
            r[1] += random_step()
            r[2] += random_step()
        r_n.append(r)                           #record the last(nth) position of our particle
        xpaths.append(xpath); ypaths.append(ypath); zpaths.append(zpath);
                                                #now compute average r for this iteration
        rave = (float(r[0])/float(steps), float(r[1])/float(steps), float(r[2])/float(steps)) 
        rave_total[0] += rave[0]                #compute running average
        rave_total[1] += rave[1]
        rave_total[2] += rave[2]
        
    rave_total[0] = float(rave_total[0])/float(iterations)       #compute averaged average position
    rave_total[1] = float(rave_total[1])/float(iterations)
    rave_total[2] = float(rave_total[2])/float(iterations)
    
    print 'Average position for all iterations:', rave_total     #display averaged avreage position
    
    t = np.arange(0, steps, 1)                                   #plot the path of the particle against time note that it
    colors = []                                                  #moves a unit step along each position vector per unit time  
    for i in range(iterations):                                   
        colors.append(numpy.random.rand(3,1))                    #generate a color for each path
        
    #for i in range(iterations):
    #    plt.plot(t, xpaths[i], color = colors[i])                #plot all paths in x direction
    #plt.show()
    #for i in range(iterations):
    #    plt.plot(t, ypaths[i], color = colors[i])                #plot all paths in y direction
    #plt.show()
    #for i in range(iterations):
    #    plt.plot(t, zpaths[i], color = colors[i])                #plot all paths in z direction
    #plt.show()
    
    #for i in range(iterations):                                  #plot all paths in the x-y plane
    #    plt.plot(xpaths[i], ypaths[i], color = colors[i])
    #plt.show()
    #for i in range(iterations):                                  #plot all paths in the y-z plane
    #    plt.plot(ypaths[i], zpaths[i], color = colors[i])
    #plt.show()
    #for i in range(iterations):                                  #plot all paths in the z-x plane
    #    plt.plot(zpaths[i], xpaths[i], color = colors[i])
    #plt.show()
    
    
    fig = pylab.figure(); ax = Axes3D(fig)
    for i in range(iterations):
        ax.plot(xpaths[i], ypaths[i], zpaths[i], color = colors[i])
    plt.show()
```
