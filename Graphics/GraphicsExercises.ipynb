(b) Now add code to calculate the running average of the data. Plot both the original data and the running average on the same graph, only over the range covered by the first 1000 data points.

```python
def running_average(data):
    average = []
    total = 0
    count = 0
    for numbers in data:
        total = total + numbers
        count = count + 1
        average.append(total / count)
    return average

z = running_average(y)


x, y = loadtxt('//Users/anthonybardessono/Desktop/sunspots.txt', unpack = True)
plt.figure(figsize = (12, 6), dpi = 180)
plt.xlim(0, 1000)
plt.ylim(0, 300)

plt.xticks([12, 312, 612, 912], 
           [r'1750', r'1775', r'1800', r'1825'])

plt.xlabel("Year", fontsize = 20)
plt.ylabel("Sunspots", fontsize = 20)


plt.plot(x[:1000], y[:1000], '^', markersize = 1.5, label = "Total Sightings", color = (0, 0, 0))
plt.plot(z[:1000], label = "Average Sightings", color = 'g')
plt.legend(loc = "upper left")
plt.title("First 1000 Months", fontsize = 23)

plt.show()
```

Make a 3D plot of the function G(x,y)=e^(-(x^2) −(y^2)) over the range -6 <(x,y)< 6 using 51 bins in x and 51 bins in y.

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')
X = np.arange(-6, 6, .25)
Y = np.arange(-6, 6, .25)
#Re-factor the x,y arrays to be a 2-D grid
X, Y = np.meshgrid(X, Y)
R = (-(X**2) - (Y**2))
Z = e**(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet,
        linewidth=0, antialiased=False)
ax.set_zlim3d(-1.01, 1.01);
```
