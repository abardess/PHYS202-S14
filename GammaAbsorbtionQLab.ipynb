```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#lead
plates = [0, 5.4, 10.6, 15.99, 21.27, 26.48, 31.71, 42.07]
N_gross = [9932, 7277, 5359, 3771, 2796, 2031, 1430, 805]
N_net = [7664, 5849, 4435, 2847, 2334, 1737, 1136, 721]
percent_uncer = [1.44, 1.60, 1.79, 2.41, 2.45, 2.78, 3.66, 4.14]

N_back = []
uncer = []
for i in range(8):
    N_back.append(N_gross[i] - N_net[i])
    uncer.append(N_net[i]*percent_uncer[i])

logN_net = []
del_logN_net = []
for i in range(8):
    logN_net.append(np.log(N_net[i]))
    del_logN_net.append(uncer[i]/N_net[i])
    
plt.errorbar(plates, logN_net, del_logN_net, 0, '.')
plt.xlabel('Total Height of Lead Plates (mm)')
plt.ylabel('Log(Net Counts) (Counts per 200 seconds)')
plt.show()
```

```python
#alluminum
aplates = [0, 13.14, 26.28, 39.42, 52.56, 65.70, 78.84, 105.12, 118.26]
aN_gross = [9915, 8317, 6874, 5731, 4639, 3828, 3135, 2222, 1811]
aN_net = [7521, 6637, 5236, 4765, 3967, 2904, 2421, 1676, 1265]
apercent_uncer = [1.48, 1.51, 1.76, 1.72, 1.84, 2.37, 2.56, 3.14, 3.84]

aN_back = []
auncer = []
for i in range(9):
    aN_back.append(aN_gross[i] - aN_net[i])
    auncer.append(aN_net[i]*apercent_uncer[i])

alogN_net = []
adel_logN_net = []
for i in range(9):
    alogN_net.append(np.log(aN_net[i]))
    adel_logN_net.append(auncer[i]/aN_net[i])
    
plt.errorbar(aplates, alogN_net, adel_logN_net, 0, '.r')
plt.xlabel('Total Height of Alluminum Plates (mm)')
plt.ylabel('Log(Net Counts) (Counts per 200 seconds)')
plt.show()
```
